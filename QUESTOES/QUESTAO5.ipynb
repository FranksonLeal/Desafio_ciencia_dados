```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leitura dos arquivos CSV

df_pagamentos = pd.read_csv('..\\BASE_DADOS\\olist_order_payments_dataset.csv')
```

```python

pagamentos = pd.DataFrame(df_pagamentos)

# Contagem da quantidade de cada tipo de pagamento
payment_counts = pagamentos['payment_type'].value_counts()
payment_counts = pd.DataFrame(payment_counts)

# RESULTADOS


# Gr√°fico de barras
payment_counts.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Quantidade de Cada Tipo de Pagamento')
plt.xlabel('Tipo de Pagamento')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()


print("Quantidade de cada categoria de pagamentos")
payment_counts


```
